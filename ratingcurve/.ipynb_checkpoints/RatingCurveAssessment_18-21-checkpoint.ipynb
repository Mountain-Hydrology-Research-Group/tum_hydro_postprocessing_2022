## Define Utility Functions
**Analyst TODO:** Run cells

```python
def plot_rating_curve(ax, site, hrange):
    ax.set_title('Rating Curve for ' + config.SITE_LONGNAME[site])
    ax.set_ylabel('Stage (m)')
    ax.set_xlabel('Q (m^3s^-1)')
    
    h = np.linspace(hrange[0], hrange[1], 100)
    Q = [config.RATING_CURVES[site](hi) for hi in h]
    
    ax.plot(Q, h, label='rating curve')
    ax.legend()
```

```python
def plot_site_stageQ(ax, site, stageQ, offset=0):
    site_stageQ = stageQ.loc[stageQ['Site Name'] == config.SITE_SHORTNAME[site]]    
    ax.scatter(site_stageQ['Measured Q (cms)'], site_stageQ['Stage (m)'] - offset, c='black', label='manual measurements')
    ax.legend()
```

### Lyell Below Maclure

It was unclear if the divergence from the rating curve at high flows represents a change in stream morphology or the natural spread of data. Plot value from previous years to evaluate.

```python
# load previous years data to evaluate spread of data relative to start points from recent years
LyellBlwMaclure_fn = 'Lyell_blw_Maclure_manual_measurements_2003_2018.csv'
LyellBlwMaclure_stageQ_pre2019_path = os.path.join('..', 'compiled_data', 'stageQ', LyellBlwMaclure_fn)

LyellBlwMaclure_pre2019_df = pd.read_csv(LyellBlwMaclure_stageQ_pre2019_path, 
                                        dtype={'stage_manually_measured(ft)': np.float64,
                                               'discharge_manual(cfs)': np.float64})

LyellBlwMaclure_pre2019_df = LyellBlwMaclure_pre2019_df.rename(columns={'stage_manually_measured(ft)':'Stage (m)'})
LyellBlwMaclure_pre2019_df['Stage (m)'] *= FT_TO_M

LyellBlwMaclure_pre2019_df = LyellBlwMaclure_pre2019_df.rename(columns={'discharge_manual(cfs)' :'Measured Q (cms)'})
LyellBlwMaclure_pre2019_df['Measured Q (cms)'] *= CFS_TO_CMS


LyellBlwMaclure_pre2019_df
```

# Vented Data Post Processing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta
import sys
import os

sys.path.insert(0, os.path.abspath(os.path.join('..', '..', 'src')))

import config
import level_baro_utils

sys.path.remove(os.path.abspath(os.path.join('..', '..', 'src')))
```

```python
sns.set_theme()
```

## Specify Site and Data Path

```python
sitecode = 2

vented_fn = 'DanaFk._297_T15min_20210707.txt'
```

## Load Data

```python
vented_path = os.path.join('..', 'data', 'raw', vented_fn)
```

```python
vented_df = pd.read_csv(vented_path, 
                        header=1, 
                        skiprows=[2,3], 
                        index_col=0, 
                        parse_dates=[0], 
                        infer_datetime_format=True)

vented_df.head()
```

```python
vented_df.columns
```

## Define Name of Level Field

```python
original_lvl_field_name = 'Lvl_ft'
new_lvl_field_name = 'lvl_cm'
```

## Convert to Correct Time Zone and Units

```python
# convert ft to cm and local time to UTC
vented_df[original_lvl_field_name] *= level_baro_utils.FT_TO_CM
vented_df.rename(columns={original_lvl_field_name:new_lvl_field_name}, inplace=True)
vented_df.sort_index(inplace=True)
vented_df.index = vented_df.index + timedelta(hours=7)
```

```python
vented_df.head()
```

## Resample (go from 15 minute data to 30 minute data)

```python
vented_df = level_baro_utils.resample_vented(vented_df)
vented_df.head()
```

## Define the Time Range to Extract Data From and Post-Process

```python
start = '2018-06'
span = '18-21'
```

```python
site_info = {'sitecode' : sitecode, 
             'site' : config.SITE_LONGNAME[sitecode], 
             'span' : span}
```

```python
vented_range_df = vented_df[start:]
vented_range_ds = vented_range_df[new_lvl_field_name]
```

## Isolate Temperature Data

```python
temp_field_name = 'Temp_C'
temp_C_ds = vented_range_df[temp_field_name]
temp_C_ds.plot(ylabel='cm')
```

## Inspect Time Range of Interest

```python
level_baro_utils.plot_multiple_pres_timeseries([vented_range_ds], [''], site_info)
```

## Inspect and Flag Ice Jams

```python
candidate_1 = {'start':'2018-11-29', 'end':'2018-12-7'}
level_baro_utils.plot_candidate(vented_range_ds, candidate_1, site_info)
```

**Verdict**: Exponential Decay and magnitude suggest rain event (same for unvented). No need to flag.

```python
candidate_2 = {'start':'2019-1-18', 'end':'2019-1-26 12:00'}
level_baro_utils.plot_candidate(vented_range_ds, candidate_2, site_info)
```

**Verdict**: Broad top and duration suggest ice jam. Flag.

```python
candidate_3 = {'start':'2019-2-5 12:00', 'end':'2019-2-26'}
level_baro_utils.plot_candidate(vented_range_ds, candidate_3, site_info)
```

**Verdict**: Ice Jam. Same peak observed for unvented. Flag. Confirmed by Jessica.

```python
candidate_4 = {'start':'2019-4-20', 'end':'2020-10-31'}
level_baro_utils.plot_candidate(vented_range_ds, candidate_4, site_info)
```

**Verdict**: Sensor malfunction. Flag.

```python
candidate_5 = {'start':'2020-12-11 14:00', 'end':'2020-12-12 0:00'}
level_baro_utils.plot_candidate(vented_range_ds, candidate_5, site_info)
```

**Verdict**: Ice jam. Flag.

```python
candidate_6 = {'start':'2020-12-12 0:00', 'end':'2021-02-01'}
level_baro_utils.plot_candidate(vented_range_ds, candidate_6, site_info)
```

**Verdict**: Ice jame observed for this period in unvented data. Flag.

```python
candidate_7 = {'start':'2021-4-01 12:00', 'end':'2021-4-08 0:00'}
level_baro_utils.plot_candidate(vented_range_ds, candidate_7, site_info)
```

**Verdict**: Initial analyis: Time lapse imagery shows melt freeze cycles. Big melt on 4-6. By 4-10 much of the surface snow has melted. Met Data inconclusive. Jessica thinks melt. Later analysis: when performing discharge QC Dana + Lyell had a spike during this period while Tuolumne did not (in particular Dana has a spike during this period while the others do not). Conclude this is an ice jam. 

```python
ice_jams = [candidate_2, candidate_3, candidate_5, candidate_6, candidate_7]
sensor_malfunctions = [candidate_4]
sensor_shifts = None
```

## Create New Frame With All Desired Output Variables

```python
output_df = level_baro_utils.make_vented_output_df(vented_range_ds.index, 
                                                   vented_range_ds, 
                                                   temp_C_ds, 
                                                   ice_jams, 
                                                   sensor_malfunctions)
```

## Split Series at Sensor Shifts and Save Data

```python
output_dfs = level_baro_utils.split_series_at_shifts(output_df, sensor_shifts)
print(f'Number of segments: {len(output_dfs)}')
```

## Save Output Data

```python
level_baro_utils.save_vented_timeseries(output_dfs, site_info)
```
